```python
# Function to convert SRT to TXT
def convert_srt_to_txt(srt_file, txt_file):
    with open(srt_file, 'r') as srt:
        with open(txt_file, 'w') as txt:
            for line in srt:
                # Remove the sequence numbers in SRT files (e.g., 1, 2, 3,...)
                if line.strip().isdigit():
                    continue
                # Write other lines (timestamps and text) to the TXT file
                txt.write(line)

# Specify the file names
srt_file = "Converted_PB_Meeting_2024_02_01_Audio.srt"  # Replace with the actual .srt file generated
txt_file = "PB_Meeting_2024_02_01_Audio.txt"

# Convert the SRT to TXT with timestamps
convert_srt_to_txt(srt_file, txt_file)

# Verify the output by reading the new TXT file
with open(txt_file, 'r') as file:
    print(file.read())
```
